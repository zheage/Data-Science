# Sistemas Baseados em Conhecimento (SBC)

São modelos que buscam resolver problemas através de conhecimento adquirido por meio de experiências anteriores, comuns quando a formulação genérica é computacionalmente complexa ou existe uma grande quantidade de conhecimento específico a ser considerado.

Sua construção é interessante quando:

- Há falta ou dispersão dee especialistas a serem consultados;
- Campo do problema é bem delimitado;
- Conhecimento pode ser facilmente perdido devido a grande quantidade de informações.



Como propriedades do SBC temos:

- Todo o conhecimento do problema deve estar explicitado em uma base, chamada de **base de conhecimento**;
- A base de conhecimento deve ser interpretada por um **mecanismo de inferência**;
- Problemas resolvidos desta forma são aqueles onde não é conhecido um procedimento determinístico, sendo necessária uma busca **heurística**.

# Sistemas especialistas

- São **sistemas baseados em conhecimentos** que são capazes de oferecer soluções para **problemas específicos** em uma determinada área de forma que seja tão bom ou melhor que de especialistas no campo.



Os componentes de um SBC consistem de:

- **Motor de inferência:** Núcleo do sistema, responsável por interpretar a base de conhecimento e realizar a inferência;
- **Memória de trabalho:** Armazena possíveis soluções, intermediárias e as soluções finais do problema;
- **Módulo explicativo:** Justificador das soluções encontradas pelo sistema;
- **Interface:** Processa a comunicação entre usuário e o SBC, onde a coleta de dados para o problema é realizada;
- **Solução recomendada:** Resposta do SBC para o usuário;
- **Usuário:** Diferentes tipos de usuários podem utilizar o SBC, como especialistas, não especialistas e outros sistemas.
  - **Modo consultor:** Para usuários leigos;
  - **Modo instrutor:** Para estudantes que desejem aprender com o SBC;
  - **Modo cooperativo:** Para auxiliar um especialista na área.

SBCs são métodos bastante utilizados em áreas como medicina, engenharia, finanças, etc. Porém, estes sofrem com dificuldades em relação a necessidade dee aquisição de conhecimento de especialistas. Além disso, muito do conhecimento pelos especialistas é difícil de ser estruturado, o que dificulta a construção dee um SBC.

Dessa forma, é comum utilizar o **Raciocínio Baseado em Casos** na hora de construir uma inferência, onde para cada novo caso, o sistema deve buscar um caso similar na base de conhecimento e utilizar a solução encontrada para o caso similar para o novo caso.
