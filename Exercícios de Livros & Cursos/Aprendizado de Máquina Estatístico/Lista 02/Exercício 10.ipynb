# Modelagem

O exercício nos pede para elaborar os seguintes classificadores:

- Regressão Logística
- Naive Bayes
- KNN

```python
X = Spam.iloc[:, :-1]
y = Spam['spam']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)

# Modelos utilizados

clf_LR = LogisticRegression()
clf_NB = GaussianNB()
clf_KNN = KNeighborsClassifier()

# Treinamento dos Modelos

clf_LR.fit(X_train, y_train)
clf_NB.fit(X_train, y_train)
clf_KNN.fit(X_train, y_train)

# Predições dos Modelos

y_predLR = clf_LR.predict(X_test)
y_predNB = clf_NB.predict(X_test)
y_predKNN = clf_KNN.predict(X_test)
```

# Métricas

```python
print(('-' * 50) + ' Regressão Logística ' + ('-' * 50))  

print(
    classification_report(y_test, y_predLR)
)

print(('-' * 50) + ' Naive Bayes ' + ('-' * 50))  

print(
    classification_report(y_test, y_predNB)
)


print(('-' * 50) + ' KNN ' + ('-' * 50))  

print(
    classification_report(y_test, y_predKNN)
)
```
