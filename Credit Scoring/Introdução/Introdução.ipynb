# Fundamentos de Credit Scoring

- A concessão de crédito é uma decisão onde sempre há a possibilidade de **perda**. Quanto há a possibilidade em estimar-se esta probabilidade, a decisão será mais confiável
- Tem como objetivo prever, na data dda decisão do crédito, a probabilidade de que caso concedido, incorra em perda ao credor;

$$ \text{Risco de crédito} = \text{Probabilidade de perda}$$

- Escores de crédito têm a finalidade única de quantificar o risco de crédito;
- A implantação e gestão de um processo automatizado de análise e decisão de crédito requer a geração e atualização de uma base de dados dos solicitantes, operações realizadas e registros de pagamentos, de forma que possamos:
  - Rever a adequação dos limites de crédito dados aos clientes;
  - Avaliar a intensidade e forma de uso de determinado produto de crédito;
  - Analisar/monitorar o perfil dos clientes potenciais do credor para orientar estratégias de marketing, como segmentação do mercado ou oferta de novos produtos de crédito

- **Risco cliente:** informações do solicitante de crédito;
- **Credit scoring x Rating:** Conceitos distintos. O primeiro é um processo quantitativo, enquanto o segundo depende grande parte de avaliação subjetiva, podendo ou não contemplar métodos quantitativos como parte do processo. 

Medir o risco de crédito de forma objetiva, utilizando técnicas quantitativas apresenta uma série de vantagens:

- Consistência nas decisões;
- Decisões rápidas
- Decisões adequadas, tornando a capaz de precificar e taxar as operações com base na probabilidade de perda de uma transação
- Decisão à distância
- Monitoramento e administração de um portfólio de crédito

- As aplicações dos modelos de scoring é fundamental para a decisão de crédito em massa
- Todavia, quando o valor do crédito a ser engociado é suficientemente alto, a alocação de um ou mais analistas experientes para analisar a operação com um maior nível de detalhamento é a opção recomendável

Na tomada de decisão entre duas alternativas, podem-se cometer dois tipos de erro:

- Recusa de solicitante que, apesar de ter um perfil de alto risco, honraria seu compromisso caso recebesse o crédito
- Aprovar uma operação de crédito que irá implicar em perdas no futuro



## Premissas Básicas em Modelos de Credit Scoring

Para desenvolvimento de modelos de credit scoring baseamo-nos em amostras de clientes aos quais condedemos crédito no passado. Isso significa que estamos supondo que, no futuro, quando aplicarmos esses modelos, o comportamento de novos clientes será igual ao dos clientes no passado

Futuro ~ passado
