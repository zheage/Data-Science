# PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification

# Gerando dados correlacionados para um problema de classificação
X, y = make_classification(
    n_samples = 10000,
    n_features = 100,
    n_informative = 90,
    n_redundant = 10,
    weights = [0.5],
    random_state = 42
)

X = pd.DataFrame(X, columns = ['X' + str(x) for x in range(1, 101)])

# Gerando o gráfico de correlações
plt.figure(figsize = (12,6))

sns.heatmap(
    X.corr()
)

plt.title('Mapa de calor da matriz de correlações')

plt.show()
```
