# Simulação de Dados Binários

```python
import pandas as pd
import numpy as np
from scipy.stats import logistic

np.random.seed(42)

tamanhoAmostra = 1000
nAmostras = 10

# Realiza a simulação para nAmostras
for _ in range(nAmostras):
    covariaveis = np.random.rand(tamanhoAmostra, 2) # Cria vetores de dimensão 2 no intervalo aleatório de (0,1)
    
    intercepto = 0.5
    coefVariavel_1 = 2
    coefVariavel_2 = -1.5
    
    # Calcula os log-odds para todas as amostras:
    logOdds = intercepto + coefVariavel_1 * covariaveis[:,0] + coefVariavel_2 * covariaveis[:, 1] # logOdds = log(p/(1-p))
    
    # Calcula as probabilidades usando a função logística:
    probs = logistic.cdf(logOdds)
    
    # Gere dados binários para todas as amostras comparando as probabilidades com valores aleatórios
    yBinario = (np.random.rand(tamanhoAmostra) < probs) 
    
    # Cria um DataFrame com os dados e as covariáveis
    data = {
        'X1' : covariaveis[:, 0],
        'X2' : covariaveis[:, 1],
        'Y' : yBinario
    }
    
    df = pd.DataFrame(data)

df
```
